```
# sin from numpy package compared to the sin approximated by n terms of a Tylor series
import numpy as np
import matplotlib.pyplot as plot


# Factorial function
def fact(n):
  if n<2:
        return 1
  else:
        return n*fact(n-1)


# Get x values of the sine wave
time_sin = np.arange(-3 , 3 , 0.1)

#____________________ sin from numpy package____________________#

# Amplitude of the sine wave is sine
# of a variable like time
amplitude_sinNP = np.sin(time_sin)

# Plot a sine wave usig time and amplitude
# obtained for the sine wave
plot.plot(time_sin,amplitude_sinNP)
#_______________________________________________________________#

#________ sin approximated by n terms of a Tylor series ________#
amplitude_sinT = []

for x in time_sin:
  a = 0
  for n in range(2):
    a = a + (((-1)**n)*((x**((2*n)+1)) / (fact((2*n)+1))))
  amplitude_sinT.append(a)


plot.plot(time_sin,amplitude_sinT)
#_______________________________________________________________#

# Give a title for the sine wave plot
plot.title('Sine and Tylor approx')

# Give x axis label for the sine wave plot
plot.xlabel('Time')

# Give y axis label for the sine wave plot
plot.ylabel('Amplitude = sin(time)')

plot.grid(True,which='both' )
plot.axhline(y=0,color='k')
plot.show()

# Display the sine wave
plot.show()
```

```
# plot or print a difference between approximated sin and the corresponding function from the numpy package
diff_sin = amplitude_sinT - amplitude_sinNP
plot.plot(time_sin, diff_sin)

#Give a title for the sine wave plot
plot.title('Diff. of sine and Tylor approx')

#Give x axis label for the sine wave plot
plot.xlabel('Time')

#Give y axis label for the sine wave plot
plot.ylabel('Amplitude')

plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()

# Display the sine wave
plot.show()
```

```
# print or plot sin2(x)+cos2(x)−1 for trygonometric functions approximated by n terms of a Tylor series

import numpy as np
import matplotlib.pyplot as plot

# Factorial function
def fact(n):
  if n<2:
        return 1
  else:
        return n*fact(n-1)


# Get x values of the wave
time = np.arange(-3, 3, 0.1);

amplitude_NP = ((np.sin(time))**2) + ((np.cos(time))**2) - 1
plot.plot(time, amplitude_NP)


amplitude_T = []

for x in time:
  a = 0
  b = 0
  for n in range(2):
    a = a + (((-1)**n)*((x**((2*n)+1)) / (fact((2*n)+1))))
    b = b + (((-1)**n)*((x**(2*n)) / (fact(2*n))))
  amplitude_T.append((a**2)+(b**2)-1)

plot.plot(time, amplitude_T)


plot.title('sin^2+cos^2-1 and Tylor approx')

plot.xlabel('Time')

plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()

plot.show()
```

```
diff = amplitude_T - amplitude_NP
plot.plot(time, diff)

plot.title('Tylor series sine and cosine')

plot.xlabel('Time')

plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
plot.show()
```
